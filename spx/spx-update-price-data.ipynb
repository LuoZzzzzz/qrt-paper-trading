```python
import pandas as pd

import refinitiv.data as rd
rd.open_session()
```

```python
old_spx = pd.read_pickle("./spx-past-5-years/spx_past_5_years")
old_spx_index = pd.read_pickle("./spx-past-5-years/spx_index_past_5_years")
```

```python
old_spx.head()
```

```python
import datetime

#start_date = str(old_spx.index[-1]).split(" ")[0]
#start_date = str(old_spx.index[-1].to_pydatetime().timedelta(days=1)).split(" ")[0]

start_date = str(old_spx.index[-1].to_pydatetime() + datetime.timedelta(days=1)).split(" ")[0]
end_date = str(datetime.date.today() - datetime.timedelta(days=1))
start_date, end_date
```

```python
new_spx = rd.get_history(
    universe=["0#.SPX"],
    fields=["TR.PriceClose"],
    interval="1D",
    start=start_date,
    end=end_date,
)

new_spx_index = rd.get_history(
    universe=[".SPX"],
    fields=["TR.PriceClose"],
    interval="1D",
    start=start_date,
    end=end_date,
)

column_names = list(new_spx_index.columns)
new_spx_index = new_spx_index.rename(columns={column_names[0]: "SPX"})
new_spx_index.columns.name = "Price Close"

new_spx.head()
```

```python
spx_index_df = pd.concat((old_spx_index, new_spx_index), axis=0)
spx_index_df.head()

spx_df = pd.concat((old_spx, new_spx), axis=0)
spx_df
```

```python
spx_df.to_pickle(f"spx_df_{end_date}")
spx_index_df.to_pickle(f"spx_index_df_{end_date}") 
```
